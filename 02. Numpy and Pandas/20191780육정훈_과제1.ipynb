```python
import pandas as pd
```

```python
def read_corpus(train_path):
    features = [['SENT_ID', 'TOKEN_ID', 'TOKEN', 'POS', 'HEAD', 'DEP', 'TASK_NAME', 'TEXT']]
    task_name = None
    text = None
    sent_id = 0
    data_array = []

    f = open(train_path, 'r')

    while True:
        line = f.readline()
        if not line:
            break

        line = line.strip()

        if line.startswith("# sent_id"):
            task_name = line

        elif line.startswith("# text"):
            text = line
            sent_id += 1

        elif not line:
            continue
            
        else:
            data = [sent_id] + line.split("\t")
            if data[1] == '1':
                data.extend([task_name, text])
    
            data_array.append(data)
            
    f.close()

    return pd.DataFrame(data=data_array, columns=features).fillna('-')
```

```python
train_path = 'train.conllu'
df_train = read_corpus(train_path)
```
